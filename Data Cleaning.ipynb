```python
# 1. Bar plot of posts per platform
plt.figure(figsize=(10, 6))
engagement_df['Platform'].value_counts().plot(kind='bar')
plt.title('Number of Posts per Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Posts')
plt.tight_layout()
plt.show()
```

```python
# 4. Pie chart of Audience Gender
plt.figure(figsize=(8, 8))
engagement_df['Audience Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Audience Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# 5. Heatmap of correlation between numeric columns
numeric_cols = ['Likes', 'Comments', 'Shares', 'Impressions', 'Reach', 'Engagement Rate']
plt.figure(figsize=(15, 8))
sns.heatmap(engagement_df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```

```python
# 6. Bar plot of top 10 Audience Locations
plt.figure(figsize=(12, 6))
engagement_df['Audience Location'].value_counts().nlargest(10).plot(kind='bar')
plt.title('Top 10 Audience Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 3. Box plot of Engagement Rate by Platform
plt.figure(figsize=(10, 6))
sns.boxplot(data=engagement_df, x='Platform', y='Engagement Rate')
plt.title('Engagement Rate Distribution by Platform')
plt.show()
```
